```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
s=pd.read_csv('50_Startups.csv')
s
```

```python
s.shape
```

CHECKING NULL VALUES

```python
s.isnull()
```

```python
s.isnull().sum()
```

```python
s.isnull().sum().sum()
```

FILLING NULL VALUES WITH MEAN

```python
s['R&D Spend']=s['R&D Spend'].fillna(value=s['R&D Spend'].mean())
s
```

```python
s['Administration']=s['Administration'].fillna(value=s['Administration'].mean())
s
```

```python
s['State'].unique()
```

```python
s['State'].replace({'New York':0, 'California':1, 'Florida':2},inplace=True)
```

```python
s.head()
```

```python
s.shape
```

```python
s.describe()
```

```python
sns.pairplot(s)
plt.show()
```

```python
x=s.iloc[:,:-1].values
y=s.iloc[:,-1].values
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train ,x_test, y_train , y_test= train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
